```python
import re
import matplotlib.pyplot as plt

epochs = []
val_accs = []
val_losses = []

current_epoch = None

for line in log.strip().split("\n"):
    # 1) Epoch 줄: [Epoch 3/100]
    epoch_match = re.search(r"\[Epoch\s+(\d+)/\d+\]", line)
    if epoch_match:
        current_epoch = int(epoch_match.group(1))
        continue

    # 2) Valid 줄에서 Loss와 Acc 추출
    valid_match = re.search(
        r"\[Valid\]\s*Loss:\s*([0-9.]+).*Acc:\s*([0-9.]+)", line
    )

    if valid_match and current_epoch is not None:
        loss = float(valid_match.group(1))
        acc = float(valid_match.group(2))

        epochs.append(current_epoch)
        val_losses.append(loss)
        val_accs.append(acc)

# ---- 그래프 1: Validation Loss ----
plt.figure(figsize=(8, 4))
plt.plot(epochs, val_losses, marker="o")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.title("Validation Loss over Epochs")
plt.grid(True)
plt.xticks(epochs)
plt.show()

# ---- 그래프 2: Validation Accuracy ----
plt.figure(figsize=(8, 4))
plt.plot(epochs, val_accs, marker="o")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.title("Validation Accuracy over Epochs")
plt.grid(True)
plt.xticks(epochs)
plt.show()
```

```python
import re
import matplotlib.pyplot as plt

epochs = []
val_accs = []
val_losses = []

current_epoch = None

for line in log.strip().split("\n"):
    # 1) Epoch 줄: [Epoch 3/100]
    epoch_match = re.search(r"\[Epoch\s+(\d+)/\d+\]", line)
    if epoch_match:
        current_epoch = int(epoch_match.group(1))
        continue

    # 2) Valid 줄에서 Loss와 Acc 추출
    valid_match = re.search(
        r"\[Valid\]\s*Loss:\s*([0-9.]+).*Acc:\s*([0-9.]+)", line
    )

    if valid_match and current_epoch is not None:
        loss = float(valid_match.group(1))
        acc = float(valid_match.group(2))

        epochs.append(current_epoch)
        val_losses.append(loss)
        val_accs.append(acc)

# ---- 그래프 1: Validation Loss ----
plt.figure(figsize=(8, 4))
plt.plot(epochs, val_losses, marker="o")
plt.xlabel("Epoch")
plt.ylabel("Validation Loss")
plt.title("Validation Loss over Epochs")
plt.grid(True)
plt.xticks(range(1, 1+len(epochs), 3))
plt.show()

# ---- 그래프 2: Validation Accuracy ----
plt.figure(figsize=(8, 4))
plt.plot(epochs, val_accs, marker="o")
plt.xlabel("Epoch")
plt.ylabel("Validation Accuracy")
plt.title("Validation Accuracy over Epochs")
plt.grid(True)
plt.xticks(range(1, 1+len(epochs), 3))
plt.show()
```

```python
import re
from dataclasses import dataclass
from typing import Dict, List, Optional, Tuple

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# ----------------------------
# 1) Parsing
# ----------------------------
@dataclass
class ParsedPGLog:
    epochs: pd.DataFrame          # one row per epoch
    debug_steps: pd.DataFrame     # one row per debug step
    coeffs: pd.DataFrame          # one row per "Coeffs ->" line


def _to_float(x: Optional[str]) -> Optional[float]:
    if x is None:
        return None
    try:
        return float(x)
    except Exception:
        return None


def parse_pgexplainer_log(log_text: str) -> ParsedPGLog:
    """
    Parse a PGExplainer training log (as a single string) into DataFrames.

    Extracts:
    - Epoch loss: "Epoch: N - loss:  X"
    - Debug step losses + mask ratio:
        Pred_Loss, Size_Loss, Ent_Loss, Total_Loss, Current Mask Ratio (Mean)
    - Coeffs lines:
        "Coeffs -> Size: a, Ent: b"
    - Model saved events:
        "Loss updated to X. Model saved."
    """
    # Patterns
    re_epoch = re.compile(r"Epoch:\s*(\d+)\s*-\s*loss:\s*([0-9]*\.?[0-9]+)")
    re_saved = re.compile(r"Loss updated to\s*([0-9]*\.?[0-9]+)\.\s*Model saved\.")
    re_debug = re.compile(
        r"Pred_Loss:\s*([0-9]*\.?[0-9]+)\s*\|\s*"
        r"Size_Loss:\s*([0-9]*\.?[0-9]+)\s*\|\s*"
        r"Ent_Loss:\s*([0-9]*\.?[0-9]+)\s*\|\s*"
        r"Total_Loss:\s*([0-9]*\.?[0-9]+)"
    )
    re_mask = re.compile(r"Current Mask Ratio \(Mean\):\s*([0-9]*\.?[0-9]+)")
    re_coeffs = re.compile(r"Coeffs\s*->\s*Size:\s*([0-9]*\.?[0-9]+),\s*Ent:\s*([0-9]*\.?[0-9]+)")

    # We'll scan line-by-line and keep "current epoch" context.
    current_epoch: Optional[int] = None

    epoch_rows: List[Dict] = []
    debug_rows: List[Dict] = []
    coeff_rows: List[Dict] = []

    pending_debug: Optional[Dict] = None  # store debug loss line until we see mask ratio line (usually next)

    lines = log_text.splitlines()
    for i, line in enumerate(lines):
        line = line.strip()

        # Coeffs lines
        m = re_coeffs.search(line)
        if m:
            coeff_rows.append({
                "epoch_context": current_epoch,
                "size_coeff": _to_float(m.group(1)),
                "ent_coeff": _to_float(m.group(2)),
                "line_idx": i,
            })
            continue

        # Debug loss line
        m = re_debug.search(line)
        if m:
            pending_debug = {
                "epoch_context": current_epoch,
                "pred_loss": _to_float(m.group(1)),
                "size_loss": _to_float(m.group(2)),
                "ent_loss": _to_float(m.group(3)),
                "total_loss": _to_float(m.group(4)),
                "mask_ratio_mean": None,
                "line_idx": i,
            }
            # We’ll try to capture mask ratio from subsequent lines (often right after).
            continue

        # Mask ratio line (attach to latest pending debug if any)
        m = re_mask.search(line)
        if m and pending_debug is not None and pending_debug.get("mask_ratio_mean") is None:
            pending_debug["mask_ratio_mean"] = _to_float(m.group(1))
            debug_rows.append(pending_debug)
            pending_debug = None
            continue

        # Epoch line
        m = re_epoch.search(line)
        if m:
            current_epoch = int(m.group(1))
            epoch_rows.append({
                "epoch": current_epoch,
                "epoch_loss": _to_float(m.group(2)),
                "line_idx": i,
                "saved_loss": None,
                "model_saved": False,
            })
            continue

        # Model saved line (attach to the most recent epoch row if possible)
        m = re_saved.search(line)
        if m:
            saved_loss = _to_float(m.group(1))
            if epoch_rows:
                epoch_rows[-1]["saved_loss"] = saved_loss
                epoch_rows[-1]["model_saved"] = True
            continue

    # If a debug line existed but mask ratio line never came, still keep it.
    if pending_debug is not None:
        debug_rows.append(pending_debug)

    epochs_df = pd.DataFrame(epoch_rows)
    debug_df = pd.DataFrame(debug_rows)
    coeffs_df = pd.DataFrame(coeff_rows)

    # Sort & clean
    if not epochs_df.empty:
        epochs_df = epochs_df.sort_values("epoch").reset_index(drop=True)

    # If epoch_context is missing (None), try to forward-fill based on order.
    # This is useful because debug prints often occur before "Epoch: N" line shows up.
    if not debug_df.empty:
        # Heuristic: map debug steps to nearest previous epoch number if available.
        # If current_epoch was None for many lines early, this stays NaN; that's OK.
        debug_df["epoch_context"] = pd.to_numeric(debug_df["epoch_context"], errors="coerce")
        debug_df = debug_df.sort_values("line_idx").reset_index(drop=True)

    if not coeffs_df.empty:
        coeffs_df["epoch_context"] = pd.to_numeric(coeffs_df["epoch_context"], errors="coerce")
        coeffs_df = coeffs_df.sort_values("line_idx").reset_index(drop=True)

    return ParsedPGLog(epochs=epochs_df, debug_steps=debug_df, coeffs=coeffs_df)


# ----------------------------
# 2) Aggregation helpers
# ----------------------------
def summarize_debug_by_epoch(debug_df: pd.DataFrame) -> pd.DataFrame:
    """
    Aggregate debug-step metrics per epoch_context (mean/median/min/max).
    If epoch_context is missing, those rows are ignored.
    """
    if debug_df is None or debug_df.empty:
        return pd.DataFrame()

    df = debug_df.copy()
    df = df.dropna(subset=["epoch_context"])
    if df.empty:
        return pd.DataFrame()

    df["epoch_context"] = df["epoch_context"].astype(int)

    agg_cols = ["pred_loss", "size_loss", "ent_loss", "total_loss", "mask_ratio_mean"]
    agg = df.groupby("epoch_context")[agg_cols].agg(["mean", "median", "min", "max", "count"])
    # Flatten columns
    agg.columns = [f"{c0}_{c1}" for (c0, c1) in agg.columns]
    agg = agg.reset_index().rename(columns={"epoch_context": "epoch"})
    return agg


# ----------------------------
# 3) Plotting
# ----------------------------
def plot_pgexplainer_training(parsed: ParsedPGLog, *,
                              show: bool = True,
                              save_dir: Optional[str] = None,
                              prefix: str = "pgexplainer") -> Dict[str, plt.Figure]:
    """
    Create key training visualizations.

    Figures:
    - Epoch loss over time (+ saved loss markers)
    - Debug-step: total_loss mean/median per epoch (if epoch_context available)
    - Debug-step: mask_ratio mean/median per epoch (if available)
    - Coeffs over time (size/ent) if present
    """
    figs: Dict[str, plt.Figure] = {}

    epochs_df = parsed.epochs
    debug_df = parsed.debug_steps
    coeffs_df = parsed.coeffs

    # A) Epoch loss
    if epochs_df is not None and not epochs_df.empty:
        fig = plt.figure()
        plt.plot(epochs_df["epoch"], epochs_df["epoch_loss"], marker="o", linewidth=1)
        if "saved_loss" in epochs_df.columns and epochs_df["saved_loss"].notna().any():
            saved = epochs_df.dropna(subset=["saved_loss"])
            plt.scatter(saved["epoch"], saved["saved_loss"], marker="x")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.title("PGExplainer Epoch Loss")
        plt.grid(True, alpha=0.3)
        figs["epoch_loss"] = fig

        if save_dir:
            fig.savefig(f"{save_dir.rstrip('/')}/{prefix}_epoch_loss.png", dpi=150, bbox_inches="tight")

    # Aggregate debug steps
    debug_agg = summarize_debug_by_epoch(debug_df)

    # B) Debug total loss summary per epoch
    if not debug_agg.empty:
        fig = plt.figure()
        plt.plot(debug_agg["epoch"], debug_agg["total_loss_mean"], marker="o", linewidth=1, label="mean")
        plt.plot(debug_agg["epoch"], debug_agg["total_loss_median"], marker=".", linewidth=1, label="median")
        plt.xlabel("Epoch (context)")
        plt.ylabel("Total_Loss (debug)")
        plt.title("Debug Total_Loss by Epoch (Aggregated)")
        plt.grid(True, alpha=0.3)
        plt.legend()
        figs["debug_total_loss"] = fig

        if save_dir:
            fig.savefig(f"{save_dir.rstrip('/')}/{prefix}_debug_total_loss.png", dpi=150, bbox_inches="tight")

    # C) Mask ratio summary per epoch
    if not debug_agg.empty and "mask_ratio_mean_mean" in debug_agg.columns:
        fig = plt.figure()
        plt.plot(debug_agg["epoch"], debug_agg["mask_ratio_mean_mean"], marker="o", linewidth=1, label="mean")
        plt.plot(debug_agg["epoch"], debug_agg["mask_ratio_mean_median"], marker=".", linewidth=1, label="median")
        plt.xlabel("Epoch (context)")
        plt.ylabel("Mask Ratio (Mean)")
        plt.title("Mask Ratio by Epoch (Aggregated)")
        plt.grid(True, alpha=0.3)
        plt.legend()
        figs["mask_ratio"] = fig

        if save_dir:
            fig.savefig(f"{save_dir.rstrip('/')}/{prefix}_mask_ratio.png", dpi=150, bbox_inches="tight")

    # D) Coeffs (if present)
    if coeffs_df is not None and not coeffs_df.empty:
        # Use line index as x if epoch_context is missing
        x = coeffs_df["epoch_context"].copy()
        use_line = x.isna().all()
        if use_line:
            x = coeffs_df["line_idx"]
            xlabel = "Log Line Index"
        else:
            xlabel = "Epoch (context)"

        fig = plt.figure()
        plt.plot(x, coeffs_df["size_coeff"], marker="o", linewidth=1, label="Size coeff")
        plt.plot(x, coeffs_df["ent_coeff"], marker="o", linewidth=1, label="Ent coeff")
        plt.xlabel(xlabel)
        plt.ylabel("Coefficient")
        plt.title("Regularization Coefficients Over Time")
        plt.grid(True, alpha=0.3)
        plt.legend()
        figs["coeffs"] = fig

        if save_dir:
            fig.savefig(f"{save_dir.rstrip('/')}/{prefix}_coeffs.png", dpi=150, bbox_inches="tight")

    if show:
        plt.show()

    return figs


# ----------------------------
# 4) Convenience: one-call runner
# ----------------------------
def visualize_pgexplainer_log(log_text: str, save_dir: Optional[str] = None) -> Tuple[ParsedPGLog, pd.DataFrame]:
    """
    Parses log -> plots -> returns (parsed, debug_agg).
    """
    parsed = parse_pgexplainer_log(log_text)
    debug_agg = summarize_debug_by_epoch(parsed.debug_steps)
    plot_pgexplainer_training(parsed, show=True, save_dir=save_dir)
    return parsed, debug_agg


# ----------------------------
# Example usage
# ----------------------------
# parsed, debug_agg = visualize_pgexplainer_log(LOG_STRING, save_dir=None)
# print(parsed.epochs.head())
# print(debug_agg.head())
```

```python
def split_debug_blocks(log_text: str):
    blocks = log_text.split('--- Debug Step ---')
    return blocks[1:]  # 첫 블록은 header라 버림

def parse_debug_steps_blockwise(log_text: str) -> pd.DataFrame:
    blocks = split_debug_blocks(log_text)
    rows = []

    for b in blocks:
        try:
            pred = re.search(r"Pred_Loss:\s*([0-9.]+)", b)
            size = re.search(r"Size_Loss:\s*([0-9.]+)", b)
            ent  = re.search(r"Ent_Loss:\s*([0-9.]+)", b)
            tot  = re.search(r"Total_Loss:\s*([0-9.]+)", b)
            mask = re.search(r"Current Mask Ratio \(Mean\):\s*([0-9.]+)", b)

            if all([pred, size, ent, tot, mask]):
                rows.append({
                    "pred_loss": float(pred.group(1)),
                    "size_loss": float(size.group(1)),
                    "ent_loss": float(ent.group(1)),
                    "total_loss": float(tot.group(1)),
                    "mask_ratio_mean": float(mask.group(1)),
                })
        except Exception:
            pass

    return pd.DataFrame(rows)


def plot_mask_ratio_from_log(log_text: str):
    df = parse_debug_steps_blockwise(log_text)

    if df.empty:
        print("❌ Debug blocks parsed = 0 (check log format)")
        return None

    plt.figure()
    plt.plot(df.index, df["mask_ratio_mean"], marker="o", linewidth=1)
    plt.xlabel("Debug Step Index")
    plt.ylabel("Current Mask Ratio (Mean)")
    plt.title("Mask Ratio per Debug Step")
    plt.grid(True, alpha=0.3)
    plt.show()

    return df


df_debug = plot_mask_ratio_from_log(LOG_STRING)

if df_debug is not None:
    print(df_debug.head())

```
